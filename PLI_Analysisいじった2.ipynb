```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import glob
import re
```

```python
# ここは手作業で設定
subject_name_list = ["kumakura","kim","souma","fujii","tubota","toki","mori","hochi","yamada","takei"]
phase_name_list = ["rest","practice","boredom","flow","ultra","overload"]
```

```python
# 正規表現でなんとでもなる
# /Users/miyakooti/repositories/arai_MATLAB_program/csv/5_toki/PLIdata
pathes = glob.glob("/content/arai_MATLAB_program/csv/[0-5]_*/PLIdata/PLI_*.csv")
pathes = sorted(pathes)
pathes
```

```python
df = pd.read_csv(f"{pathes[0]}")
```

```python
df.plot()
plt.title("time series variation of PLI")
plt.xlabel("seconds")
plt.ylabel("PLI")
plt.legend()
```

```python
num_of_data_per_subject = 6
num_of_subject = int(len(pathes)/num_of_data_per_subject)

fig = plt.figure(figsize=(30,30))

for i, path in enumerate(pathes):
    df = pd.read_csv(f"{path}")
    plt.ylim(0,1)
    plt.xlim(0,250)
    plt.subplot(num_of_data_per_subject,num_of_subject,i+1)
    plt.plot(df)
    subject_name = subject_name_list[i//6]
    phase_name = phase_name_list[i%6]
    plt.title(f"{subject_name}_{phase_name}")

# 　　正規表現できんかった
#     res = re.search("*_*\.", path)
#     print(path[res])

#plt.savefig("../images/PLI_of_all_subjects.png")
plt.show()

# 低い方が負荷高い酢
```
